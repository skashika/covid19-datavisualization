```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import plotly.express as px
import plotly.graph_objs as go
from plotly.subplots import make_subplots
from fbprophet.plot import plot_plotly, add_changepoints_to_plot
import plotly.offline as py
from datetime import date, timedelta
from statsmodels.tsa.arima_model import ARIMA
from sklearn.cluster import KMeans
from fbprophet import Prophet
```

```python
data1=pd.read_csv(r"C:\Users\Owner\Downloads\covid_19_data.csv")
```

```python
data1
```

```python
daily = data1.sort_values(['ObservationDate','Country/Region','Province/State'])
latest = data1[data1.ObservationDate == daily.ObservationDate.max()]
latest.head()
```

```python
data=latest.rename(columns={ "Country/Region": "country", "Province/State": "state","Confirmed":"confirm","Deaths": "death","Recovered":"recover"})
data.head()
```

```python
dgc=data.groupby("country")[['confirm', 'death', 'recover']].sum().reset_index()

dgc.head()
dgc.describe().T
```

```python
fig = px.bar(dgc[['country', 'confirm']].sort_values('confirm', ascending=False),
             y="confirm", x="country", color='country',
             log_y=True, template='ggplot2', title='Confirmed Cases')
fig.show()
fig = px.bar(dgc[['country', 'recover']].sort_values('recover', ascending=False),
             y="recover", x="country", color='country',
             log_y=True, template='ggplot2', title='Recovered Cases')
fig.show()

fig = px.bar(dgc[['country', 'death']].sort_values('death', ascending=False),
             y="death", x="country", color='country',
             log_y=True, template='ggplot2', title='Death')
fig.show()
```

```python
dates = data1.groupby("ObservationDate")[['Confirmed', 'Deaths', 'Recovered']].sum().reset_index()
dates
```

```python
dates_conf = dates.iloc[:,[0,1]]
dates_conf
```

```python
prophet=dates_conf
prophet.tail()
```

```python
prophet.columns = ['ds','y']
prophet.tail()
```

```python
model = Prophet()
model.fit(prophet)
future=model.make_future_dataframe(periods=15)
forecast=model.predict(future)
print(forecast)
```

```python
cnfrm = forecast.loc[:,['ds','trend']]
cnfrm = cnfrm[cnfrm['trend']>0]
cnfrm=cnfrm.tail(15)
cnfrm
```

```python
cnfrm.columns = ['Date','Confirm']
cnfrm.head()
```

```python
figure = plot_plotly(model, forecast)
py.iplot(figure)
```

```python
figure = model.plot(forecast,xlabel='Date',ylabel='Confirmed Count')
figure.show()
```

```python
figure=model.plot_components(forecast)
figure.show()
```

```python
prophet_rec=dates.loc[: , ["ObservationDate","Recovered" ]].copy()
prophet_rec.tail()
prophet_rec.columns = ['ds','y']
prophet_rec.head()
```

```python
m1=Prophet()
m1.fit(prophet_rec)
future_rec=m1.make_future_dataframe(periods=15)
forecast_rec=m1.predict(future_rec)
print(forecast_rec)
```

```python
rec = forecast_rec.loc[:,['ds','trend']]
rec = rec[rec['trend']>0]
rec=rec.tail(15)
rec.columns = ['Date','Recovery']
print(rec.head())
```

```python
figure_rec = plot_plotly(m1, forecast_rec)
py.iplot(figure_rec)
```

```python
figure_rec = m1.plot(forecast_rec,xlabel='Date',ylabel='Recovery Count')
figure_rec.show()
```
